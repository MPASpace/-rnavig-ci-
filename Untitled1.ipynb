```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ==========================================
# 1. LÉPÉS: ÁLLOMÁS ADATAI ÉS KONSTANSOK
# ==========================================
print("=" * 70)
print("1. FELADAT: ÁLLOMÁS ADATAINAK BEKÉRÉSE")
print("Szakmai magyarázat:")
print("A számítások a vonatkoztatási keret és a referenciaállomás stabilitása")
print("nélkül csak 'lebegnek' a térben. A hálózatba kapcsolt GNSS pontok adják")
print("meg azt a szilárd alapot, amihez minden mást viszonyíthatunk.")
print("=" * 70 + "\n")

# IDE ÍRD BE A SAJÁT ÁLLOMÁSOD ADATAIT A STATION LOG FÁJLBÓL!
X_sta = 4052449.78  # Saját X koordináta
Y_sta = 1416954.10  # Saját Y koordináta
Z_sta = 4701446.54  # Saját Z koordináta

# WGS84 ellipszoid geometriai adatai
a_wgs84 = 6378137.0
f_wgs84 = 1.0 / 298.257223563

# ==========================================
# 2. LÉPÉS: TRANSZFORMÁCIÓS FÜGGVÉNYEK
# ==========================================
print("=" * 70)
print("2. FELADAT: ECEF ÉS FÖLDRAJZI TRANSZFORMÁCIÓK")
print("Szakmai magyarázat:")
print("Az ECEF és a földrajzi koordináták közötti oda-vissza alakítás")
print("a műholdas helymeghatározás alapja. Az iterációs eljárások ")
print("mutatják meg, hogyan illeszthető rá egy tiszta matematikai")
print("ellipszoid a valós, fizikai térbeli mérésekre.")
print("=" * 70 + "\n")

def ecef2geog(X, Y, Z, a, f):
    """ECEF térbeli derékszögű -> Földrajzi koordináták (iterációval)"""
    b = a * (1 - f)
    e2 = (a**2 - b**2) / a**2
    p = np.sqrt(X**2 + Y**2)

    phi_0 = np.arctan(Z / ((1 - e2) * p))
    tol = 1e-9
    diff = np.inf
    h_0 = 0.0

    while abs(diff) > tol:
        N_1 = a / np.sqrt(1 - e2 * np.sin(phi_0)**2)
        h_1 = p / np.cos(phi_0) - N_1
        phi_1 = np.arctan(Z / (p * (1 - e2 * N_1 / (N_1 + h_1))))
        diff = h_1 - h_0
        phi_0 = phi_1
        h_0 = h_1

    lam_rad = np.arctan2(Y, X)
    return np.array([np.degrees(phi_0), np.degrees(lam_rad), h_0])

def geog2ecef(phi_dd, lam_dd, h, a, f):
    """Földrajzi -> ECEF térbeli derékszögű koordináták"""
    b = a * (1 - f)
    e2 = (a**2 - b**2) / a**2
    phi_rad, lam_rad = np.radians(phi_dd), np.radians(lam_dd)
    N = a / np.sqrt(1 - e2 * np.sin(phi_rad)**2)

    X = (N + h) * np.cos(phi_rad) * np.cos(lam_rad)
    Y = (N + h) * np.cos(phi_rad) * np.sin(lam_rad)
    Z = ((1 - e2) * N + h) * np.sin(phi_rad)
    return np.array([X, Y, Z])

def ecef2horizontal(X, Y, Z, lat_topo, lon_topo, h_topo, a, f):
    """ECEF -> Horizonti koordináták (Azimut, Magassági szög, Távolság)"""
    topo_ecef = geog2ecef(lat_topo, lon_topo, h_topo, a, f)
    range_vec = np.array([X - topo_ecef[0], Y - topo_ecef[1], Z - topo_ecef[2]])

    phi, lam = np.radians(lat_topo), np.radians(lon_topo)
    R_topo = np.array([
        [-np.sin(lam),              np.cos(lam),             0],
        [-np.sin(phi)*np.cos(lam), -np.sin(phi)*np.sin(lam), np.cos(phi)],
        [ np.cos(phi)*np.cos(lam),  np.cos(phi)*np.sin(lam), np.sin(phi)]
    ])

    enu = np.dot(R_topo, range_vec)
    e, n, u = enu[0], enu[1], enu[2]

    azimuth = np.degrees(np.arctan2(e, n))
    if azimuth < 0: azimuth += 360
    elevation = np.degrees(np.arctan2(u, np.sqrt(e**2 + n**2)))
    slant_range = np.linalg.norm(range_vec)

    return np.array([azimuth, elevation, slant_range])

# Végrehajtás és ellenőrzés
geog_coords = ecef2geog(X_sta, Y_sta, Z_sta, a_wgs84, f_wgs84)
print(f"Bemeneti ECEF (X, Y, Z): {X_sta}, {Y_sta}, {Z_sta}")
print(f"Kiszámolt Földrajzi (Szélesség, Hosszúság, Magasság): {geog_coords}")
ecef_coords = geog2ecef(geog_coords[0], geog_coords[1], geog_coords[2], a_wgs84, f_wgs84)
print(f"Visszaszámolt ECEF (ellenőrzés): {ecef_coords}\n")

# ==========================================
# 3. LÉPÉS: FELDOLGOZÁS ÉS VIZUALIZÁCIÓ
# ==========================================
print("\n" + "=" * 70)
print("3. FELADAT: TOPOCENTRIKUS RENDSZER ÉS SKYPLOT")
print("Szakmai magyarázat:")
print("A topocentrikus transzformáció során az elvont térbeli vektorokból")
print("a lokális megfigyelő horizontján mérhető, azonnal értelmezhető navigációs")
print("információ lesz. A skyplot pedig a műholdgeometria gyors vizuális ellenőrzése.")
print("=" * 70 + "\n")

# Műhold adatok az új 2025/26/2. féléves feladatból
sat_data = {
    'id': ['283', '168', '248', '311', '140', '286', '083', '057', '093', '101', '043', '257', '024', '128', '127', '251', '034', '075', '086', '182', '014', '240', '262', '247'],
    'X_km': [17572.516173, 4025.219139, 942.417130, 11741.760891, 18506.514416, 2522.945834, -6510.239959, -3160.127014, -14605.288846, -11006.197802, -10658.774715, -152.884570, -14278.780023, -1339.030205, -23973.744600, -26259.561528, -13394.622910, -324.509391, 5069.563996, 20455.989512, 3842.818124, 10667.047884, 814.613227, 8254.842596],
    'Y_km': [18965.880531, 19873.897605, 11417.315592, 14199.652500, 13095.639278, 4059.620505, 25007.268585, 1896.391944, 1672.141549, 12003.742967, 5919.220445, 64.547983, -14565.693406, -2679.513525, -2161.239782, -3251.896211, -16799.792036, -26303.949233, -878.583125, -4075.451649, -3079.261026, -3416.476053, -8104.752562, -1356.484892],
    'Z_km': [5746.143625, -17048.472493, 23790.583228, -18882.847159, 13638.247097, -26041.615481, 5797.269291, -26215.529139, 22038.269902, -20862.626434, 23596.461733, -26514.403813, 16845.344034, -26337.596564, 10913.594072, -1108.571435, 15432.594075, -2567.187737, 25877.079690, -16335.114823, 26031.516324, -24068.487117, 25279.414308, -25162.895617]
}

df = pd.DataFrame(sat_data)
visible_sats = []
lat_t, lon_t, h_t = geog_coords

print("--- LÁTHATÓ MŰHOLDAK HORIZONTI KOORDINÁTÁI ---")
for index, row in df.iterrows():
    X_m, Y_m, Z_m = row['X_km'] * 1000, row['Y_km'] * 1000, row['Z_km'] * 1000
    az, el, ran = ecef2horizontal(X_m, Y_m, Z_m, lat_t, lon_t, h_t, a_wgs84, f_wgs84)

    if el > 0: # Csak a horizont feletti műholdak
        visible_sats.append({'id': row['id'], 'azimuth': az, 'elevation': el, 'range': ran})
        print(f"Műhold ID: {row['id']} | Azimut: {az:.2f}° | Magasság: {el:.2f}° | Távolság: {ran:.2f} m")

# Skyplot rajzolása
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'}, figsize=(8, 8))
for sat in visible_sats:
    az_rad, el_plot = np.radians(sat['azimuth']), 90 - sat['elevation']
    ax.plot(az_rad, el_plot, 'o', markersize=8, color='blue')
    ax.annotate(sat['id'], xy=(az_rad, el_plot), xytext=(5, 5), textcoords='offset points', ha="left", va="bottom")

ax.set_theta_zero_location("N")
ax.set_theta_direction(-1)
ax.set_rticks([90, 60, 30, 0])
ax.set_rgrids([0, 30, 60, 90], labels=["90°", "60°", "30°", "0°"])
ax.set_rlabel_position(45)
ax.set_thetagrids(np.rad2deg(np.linspace(0, 2*np.pi, 5))[1:], labels=["E", "S", "W", "N"])

plt.title("Látható műholdak Skyplot ábrája", va='bottom', fontsize=14, fontweight='bold')
plt.show()
```
