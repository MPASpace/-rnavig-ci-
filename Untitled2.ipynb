```python
# 1. Sz眉ks茅ges k枚nyvt谩rak telep铆t茅se (Colab-ban ezt is futtatni kell)
!pip install skyfield folium

from skyfield.api import Topos, load
import folium

# 2. TLE p谩lyaadatok bet枚lt茅se a Celestrak akt铆v m疟holdas adatb谩zis谩b贸l
print("P谩lyaadatok let枚lt茅se...")
stations_url = 'https://celestrak.org/NORAD/elements/gp.php?GROUP=active&FORMAT=tle'
satellites = load.tle_file(stations_url)

# M疟hold keres茅se n茅v alapj谩n
by_name = {sat.name: sat for sat in satellites}
target_name = 'HUNITY' # Ha k茅s艖bb beker眉l a v茅gleges n茅vvel (vagy NORAD ID-val), ez alapj谩n fogja megtal谩lni

if target_name in by_name:
    satellite = by_name[target_name]
    print(f"Szuper, a(z) {target_name} megtal谩lva az adatb谩zisban!")
else:
    print(f"A(z) {target_name} n茅v m茅g nincs a Celestrak akt铆v list谩j谩ban (az adatb谩zis val贸sz铆n疟leg friss铆t茅s alatt 谩ll).")
    print("脕tmenetileg a SMOG-1 adataival demonstr谩ljuk a k枚vet艖 algoritmust...")
    satellite = by_name.get('SMOG-1', satellites[0])

# 3. Jelenlegi id艖pont lek茅r茅se
ts = load.timescale()
t = ts.now()

# 4. Poz铆ci贸 (sz茅less茅g, hossz煤s谩g, magass谩g) kisz谩m铆t谩sa
geocentric = satellite.at(t)
subpoint = geocentric.subpoint()

lat = subpoint.latitude.degrees
lon = subpoint.longitude.degrees
elevation_km = subpoint.elevation.km

print("\n" + "-" * 40)
print(f"M疟hold: {satellite.name}")
print(f"Id艖pont (UTC): {t.utc_datetime().strftime('%Y-%m-%d %H:%M:%S')}")
print(f"Sz茅less茅g: {lat:.4f}掳")
print(f"Hossz煤s谩g: {lon:.4f}掳")
print(f"Magass谩g: {elevation_km:.1f} km")
print("-" * 40 + "\n")

# 5. T茅rk茅pes megjelen铆t茅s
# L茅trehozunk egy t茅rk茅pet, ami pontosan a m疟hold al谩 f贸kusz谩l
m = folium.Map(location=[lat, lon], zoom_start=4)

# Jel枚l艖 elhelyez茅se a t茅rk茅pen
folium.Marker(
    [lat, lon],
    popup=f"<b>{satellite.name}</b><br>Magass谩g: {elevation_km:.1f} km",
    icon=folium.Icon(color="blue", icon="info-sign")
).add_to(m)

# Egy nagys谩grendileg 1000 km-es sugar煤 k枚r, ami a hozz谩vet艖leges l谩t贸hat谩rt (footprint) jel枚li
folium.Circle(
    radius=1000000,
    location=[lat, lon],
    color="blue",
    fill=True,
    fill_opacity=0.1
).add_to(m)

# T茅rk茅p megjelen铆t茅se a cella alatt
m
```
