```python
# 1. Szükséges könyvtárak telepítése (Colab-ban ezt is futtatni kell)
!pip install skyfield folium

from skyfield.api import Topos, load
import folium

# 2. TLE pályaadatok betöltése a Celestrak aktív műholdas adatbázisából
print("Pályaadatok letöltése...")
stations_url = 'https://celestrak.org/NORAD/elements/gp.php?GROUP=active&FORMAT=tle'
satellites = load.tle_file(stations_url)

# Műhold keresése név alapján
by_name = {sat.name: sat for sat in satellites}
target_name = 'HUNITY' # Ha később bekerül a végleges névvel (vagy NORAD ID-val), ez alapján fogja megtalálni

if target_name in by_name:
    satellite = by_name[target_name]
    print(f"Szuper, a(z) {target_name} megtalálva az adatbázisban!")
else:
    print(f"A(z) {target_name} név még nincs a Celestrak aktív listájában (az adatbázis valószínűleg frissítés alatt áll).")
    print("Átmenetileg a SMOG-1 adataival demonstráljuk a követő algoritmust...")
    satellite = by_name.get('SMOG-1', satellites[0])

# 3. Jelenlegi időpont lekérése
ts = load.timescale()
t = ts.now()

# 4. Pozíció (szélesség, hosszúság, magasság) kiszámítása
geocentric = satellite.at(t)
subpoint = geocentric.subpoint()

lat = subpoint.latitude.degrees
lon = subpoint.longitude.degrees
elevation_km = subpoint.elevation.km

print("\n" + "-" * 40)
print(f"Műhold: {satellite.name}")
print(f"Időpont (UTC): {t.utc_datetime().strftime('%Y-%m-%d %H:%M:%S')}")
print(f"Szélesség: {lat:.4f}°")
print(f"Hosszúság: {lon:.4f}°")
print(f"Magasság: {elevation_km:.1f} km")
print("-" * 40 + "\n")

# 5. Térképes megjelenítés
# Létrehozunk egy térképet, ami pontosan a műhold alá fókuszál
m = folium.Map(location=[lat, lon], zoom_start=4)

# Jelölő elhelyezése a térképen
folium.Marker(
    [lat, lon],
    popup=f"<b>{satellite.name}</b><br>Magasság: {elevation_km:.1f} km",
    icon=folium.Icon(color="blue", icon="info-sign")
).add_to(m)

# Egy nagyságrendileg 1000 km-es sugarú kör, ami a hozzávetőleges látóhatárt (footprint) jelöli
folium.Circle(
    radius=1000000,
    location=[lat, lon],
    color="blue",
    fill=True,
    fill_opacity=0.1
).add_to(m)

# Térkép megjelenítése a cella alatt
m
```
